### Processing the data

We are only intrested in data with type "Earthquake"

```python
dataframe = dataframe[dataframe.Type == "Earthquake"]
dataframe["Type"].value_counts()
```

We need to convert the string format to something more manegable by a neural network. While going through the data i found 2 date time formats. I chose the POSIX timestamp to represent the date and time.

```python
def datetimeconverter(row):
    try:
        return datetime.strptime(row["Time"] + " " + row["Date"],"%H:%M:%S %m/%d/%Y").timestamp()
    except ValueError:
        pass
    
    return datetime.strptime(row["Time"][:19], "%Y-%m-%dT%H:%M:%S").timestamp()
    
   
        
dataframe["DateTime"] = dataframe.apply(datetimeconverter, axis = 1)
```

Let's remove the features we are not intrested in for the challenge

```python
dataframe = dataframe[["DateTime", "Latitude", "Longitude", "Magnitude"]]
dataframe.head()
```

now we need to do something about the date and time, we have to create a function because there are 2 types of formats

## Creating the Test Set

```python
trainDF, testDF = skl.model_selection.train_test_split(dataframe, test_size = 0.2, random_state = 42)
print(trainDF.size / dataframe.size)
testDF[["Magnitude"]].hist()
```

```python
trainX = trainDF[["DateTime", "Latitude", "Longitude"]].values
trainY = trainDF[["Magnitude"]].values
testX = testDF[["DateTime", "Latitude", "Longitude"]].values
testY = testDF[["Magnitude"]].values
print("testX shape:{}, testY shape:{}\n"\
    "trainX shape:{} trainY shape:{}".format(
        testX.shape, testY.shape, trainX.shape, trainY.shape))
```

___
## Building deep neural network

```python
input_layer = tfl.input_data(shape=[None, 3])
dense1 = tfl.fully_connected(input_layer, 3, activation='tanh',
                                 regularizer='L2', weight_decay=0.001)
dropout1 = tfl.dropout(dense1, 0.8)
dense2 = tfl.fully_connected(dropout1, 10, activation='tanh',
                                 regularizer='L2', weight_decay=0.001)
dropout2 = tfl.dropout(dense2, 0.8)
softplus = tfl.fully_connected(dropout2, 1, activation='softplus')

# Regression using SGD with learning rate decay and Top-3 accuracy
network = tfl.regression(softplus, optimizer='adam',
                     loss='categorical_crossentropy',
                     learning_rate=0.001)
```

```python
#traning
model = tfl.DNN(network, tensorboard_verbose = 0)
model.fit(trainX, trainY, validation_set = (testX, testY), show_metric = True,
         batch_size = 32)
```
